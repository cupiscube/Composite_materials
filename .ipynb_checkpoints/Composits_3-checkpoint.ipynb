```python
df1 = pd.DataFrame(pd.read_excel('./data/X_bp.xlsx', index_col=0))
df2 = pd.DataFrame(pd.read_excel('./data/X_nup.xlsx', index_col=0))
df = pd.concat([df1, df2], join='inner', axis=1)
df = df[23:]
df
```

```python
df.describe()
```

```python
df['Поверхностная плотность, г/м2 _log'] = np.log(df['Поверхностная плотность, г/м2'])
```

```python
df['Количество отвердителя, м.% _exp'] = np.exp(df['Количество отвердителя, м.%'])
```

```python
df['модуль упругости, ГПа _log'] = np.log(df['модуль упругости, ГПа'])
```

```python
df['модуль упругости, ГПа _log_10'] = np.log10(df['модуль упругости, ГПа'])
```

```python
df['нашивка'] = df['Плотность нашивки'] / df['Шаг нашивки']
```

```python
df['нашивка_2'] = df['Плотность нашивки'] * df['Шаг нашивки']
```

```python
df['Потребление смолы, г/м2 _log_10'] = np.log10(df['Потребление смолы, г/м2'])
```

```python
df['Плотность, кг/м3 _г/м2'] = df['Плотность, кг/м3'] ** (2/3) / 1000
```

```python
df['Температура вспышки, С_2 нормальное'] = 2 ** df['Температура вспышки, С_2']
```

```python
df['Содержание эпоксидных групп,%_2 нормальное'] = 2 ** df['Содержание эпоксидных групп,%_2']
```

```python
df['модуль упругости, МПа'] = df['модуль упругости, ГПа'] / 1000
```

```python
df['Модуль упругости при растяжении, МПа'] = df['Модуль упругости при растяжении, ГПа'] / 1000
```
